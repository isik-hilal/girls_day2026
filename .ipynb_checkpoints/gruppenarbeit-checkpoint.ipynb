Liebe Girls 🤖💖,

arbeitet in euren Gruppen zusammen und sucht **inspirierende Frauen/Girls** aus **beliebigen Bereichen**.  
Beispiele: Informatik, Wissenschaft, Sport, Musik, Kunst, Aktivismus, YouTube / Social Media, Politik, Entrepreneurship …  

Ihr könnt die verschiedenen Versionen der Bücher **Good Night Stories for Rebel Girls** nutzen, die wir euch mitgebracht haben, um eure Recherche und euer Quiz vorzubereiten.

Eure Aufgabe:  
- Bereitet **3 Fragen** für uns vor, basierend auf den Informationen über diese Frauen.  
- Jede Frage sollte **Clues**, **Multiple-Choice-Optionen** und eine kurze Erklärung enthalten.  
- Nutzt Python, um **ein kleines Quiz** zu erstellen.  

⏰ Ihr habt **45 Minuten**, um euer Code-Projekt zu gestalten.  

Wichtig:  
- Dies ist **nicht nur Codieren** – ihr übt auch **Recherche, Teamarbeit und Präsentation**.

Bonus (optional, keine Pflicht):  
- Fügt Pausen zwischen den Fragen ein (`time.sleep`)  
- Passt die Größe der Bilder an (`width`, `height`)  
- Kreative Extras, die euer Quiz spannender machen  

Am Ende präsentiert jede Gruppe ihren Code, und wir alle versuchen, die Fragen zu beantworten.  

Viel Spaß & viel Erfolg! 🚀


# Starter-Canvas: Girl Power Quiz
## Jede Gruppe füllt die Lücken aus

```python
from IPython.display import Image, display
import time  # Bonus: für Pausen zwischen den Fragen

# -------------------------------
# Runde 1: 
# -------------------------------

print("🎉 Rate die inspirierende Frau! 🎉")
print("\nRunde 1")

# Hinweise (hier Lücken ausfüllen)
print("\nHinweise:")
print("1. ...")
time.sleep(1)
print("2. ...")
time.sleep(1)
print("3. ...\n")

# Multiple Choice (hier eure Optionen eintragen)
print("Wer ist sie?")
print("A: ...")
print("B: ...")
print("C: ...")

antwort = input("Deine Antwort (A/B/C): ").upper()

if antwort == "A":
    print("✅ Richtig! Es ist ...")
else:
    print("❌ Falsch! Die richtige Antwort ist ...")

# Warum sie wichtig ist (hier ergänzen)
print("\nWarum sie wichtig ist:")
print("👉 ...")
print("👉 ...\n")

# Bild einfügen (Bonus: width und height anpassen)
display(Image(url="...", width=200, height=200))
```
