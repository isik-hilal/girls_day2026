```python
from IPython.display import Image, display

print("🎉 Rate die inspirierende Frau! 🎉")
print("\nRunde 1: Historische Computerpionierin")

print("\nHinweise:")
print("1. Sie gilt als erste Programmiererin.")
print("2. Sie schrieb einen Algorithmus für Charles Babbages Maschine.")
print("3. Sie lebte im 19. Jahrhundert.\n")

print("Wer ist sie?")
print("A: Ada Lovelace")
print("B: Grace Hopper")
print("C: Marie Curie")

antwort = input("Deine Antwort (A/B/C): ").upper()

if antwort == "A":
    print("✅ Richtig! Es ist Ada Lovelace!")
else:
    print("❌ Falsch! Die richtige Antwort ist Ada Lovelace.")

print("\nWarum sie wichtig ist:")
print("👉 Sie stellte sich Computer vor, die mehr als nur rechnen konnten.")
print("👉 Sie schrieb den ersten Algorithmus für eine Maschine.\n")

display(Image(url="https://upload.wikimedia.org/wikipedia/commons/a/a4/Ada_Lovelace_portrait.jpg", width=300))

# Infos zu den anderen Antworten
print("\nKurze Infos zu den anderen Frauen:")
print("Grace Hopper: US-amerikanische Informatikerin, arbeitete an frühen Programmiersprachen.")
print("Marie Curie: Polnisch-französische Wissenschaftlerin, bekannt für Forschung zur Radioaktivität.\n")

print("💡 Motivation: Wenn Ada programmieren konnte, kannst du es auch! 💻👩‍💻")
```
