# Python Schritt-für-Schritt: `input()`

## Schritt 1: Was ist input()?

`input()` ist eine Funktion in Python, die es erlaubt, vom Benutzer etwas einzugeben.
Die Eingabe wird als Text (String) gespeichert.

```python
name = input("Wie heißt du? ")
print("Hallo,", name, "👋")
```

```python
hobby = input("Was ist dein Hobby? ")
print("Dein Hobby ist:", hobby)
```

```python
app = input("Welche App benutzt du am liebsten? ")
print("Deine Lieblingsapp ist:", app)
```

## Du bist dran!


```python
media = input("Was ist deine Lieblingsserie, Film oder Buch? ")
print("Dein Favorit ist:", media)
```

## If/else

```python
media = input("Wer ist deine Lieblings-Rapperin? ")

if media.lower() == "mimi":
    print("Cool! Du liebst mimi 🎤")
else:
    print("Ah, dein Favorit ist:", media, "– Super Wahl! 👍")
```

## Mini Quiz Beispiel

```python
print("Wer ist deine Lieblings-Rapperin?")
print("A: Nina Chuba")
print("B: Mimi")
print("C: Shirin David")

antwort = input("Deine Antwort (A/B/C): ").upper()

if antwort == "A":
    print("Cool! Deine Lieblings-Rapperin ist Nina Chuba 🎵")
elif antwort == "B":
    print("Cool! Deine Lieblings-Rapperin ist Mimi 🎤")
elif antwort == "C":
    print("Cool! Deine Lieblings-Rapperin ist Shirin David 🎶")
else:
    print("Hmm, diese Antwort kenne ich nicht 😅")
```
